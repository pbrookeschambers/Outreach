```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
# fig = plt.figure(aspect = 1, figsize = (8, 8))
for i in range(3):
    theta = np.linspace(0, 10*2*np.pi, n) * (-1) ** i
    f = np.sin(np.linspace(0, 2*np.pi, n) * np.random.random() * 2 * np.linspace(1, 1.25, n)) + np.random.random() * 2
    r = np.cos(theta * f + np.random.random() + 1) + f
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.plot(x, y, alpha = 0.75, color = "#3E3E3F")
    
    plt.axis("off")
    plt.gca().set_aspect('equal')
plt.savefig("random_scribble.png", dpi = 300, bbox_inches = 'tight', pad_inches = 0)
plt.show()
```
